###  Combinación del dataframe Accidentes con dataframe de Eventos meteorológicos.
Se ha decidido agrupar los accidentes y eventos por estado y día del suceso.
Además se va a añadir los subtotales del grado de severidad de accidente. Lo mismo se va a hacer con el dataframe de eventos meteorologicos, manteniendo subtotales de severidad meteorologica y tipo de evento evento meteorologico.

Se crea la columna date de ambos dataframe

```python
df_acc_year["Date"] = df_acc_year["Start_Time"].dt.date
df_weather_year["Date"] = df_weather_year["StartTime(UTC)"].dt.date
```

Contamos los eventos meteorologicos por estado y día.

```python
weather_tot_events = (
    df_weather_year
    .groupby(["State", "Date"])
    .size()
    .reset_index(name="weather_events")
)
```

Añadimos cuenta con el tipo de evento meteorologico

```python
weather_by_type = (
    df_weather_year
    .groupby(["State","Date", "Type"])
    .size()
    .unstack(fill_value=0)
    .add_prefix("typ_")
    .reset_index()
)
```

Añadimos cuenta con la severidad

```python
weather_by_severity = (
    df_weather_year
    .groupby(["State", "Date", "Severity"])
    .size()
    .unstack(fill_value=0)
    .add_prefix("wea_sev_")
    .reset_index()
)
```

Lo unificamos todo en un data frame

```python
weather_agg = (
    weather_tot_events
    .merge(weather_by_type, on=["State", "Date"], how="left")
    .merge(weather_by_severity, on=["State", "Date"], how="left")
)
```

```python
se.eda_prelim(weather_agg)
weather_agg.sample(5)
```

Vamos a agrupar el numero total de accidentes por día y estado añadiendo el subtotal segun la severidad del accidente.

```python
acc_tot = (
    df_acc_year
    .groupby(["State", "Date"])
    .size()
    .reset_index(name="acc_events"))
```

```python
acc_by_severity = (
    df_acc_year
    .groupby(["State", "Date", "Severity"])
    .size()
    .unstack(fill_value=0)
    .add_prefix("acc_sev_")
    .reset_index())
```

Unimos ambos

```python
acc_agg = (
    acc_tot
    .merge(acc_by_severity, on=["State", "Date"], how="left")
)
```

```python
se.eda_prelim(acc_agg)
acc_agg.sample(5)
```

Creamos nuestro dataframe final con el agregado de accidentes con eventos meteorologicos

```python
df_final = acc_agg.merge(
    weather_agg,
    on=["State", "Date"],
    how="outer"
)
```

Se van a añadir los nombres de los estados de estados unidos para tener más claro los datos

```python
df_states = pd.read_csv('../data/raw/United States - Region Codes.csv', low_memory=False)
df_states.sample(5)
```

VAmos a cambiar los nombres de las columnas para hacer el merge y a estandarizar los registros de los codigos de estado con el fin de hacerlos conicidir con el df_final poniendolos en mayúsculas y eliminando el "US-"

```python
df_states= df_states.rename(columns={"Region code": "State","Region":"Name"})
df_states["State"] = df_states["State"].str.upper()
df_states["State"] = df_states["State"].str.replace("US-", "", regex=False)
df_states.sample(5)
```

Introducimos en nuestro df_final el nombre del estado para una mayor comprension de los datos

```python
df_final = df_final.merge(
    df_states[["State", "Name"]],
    on="State",
    how="left")
```

```python
se.eda_prelim(df_final)
```

Guardamos nuestro nuevo archivo con ambos archivos juntos y pasamos a la limpieza

```python
df_final.to_csv('../data/Acc_met.csv', index=False)
```
